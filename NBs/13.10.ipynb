```python
import snoop
```

```python
%load_ext blackmagic
```

```python
@snoop
def division_numeral(num1, num2, power):
    digits = list(map(int, list(str(num1))))
    repr = []
    i = len(digits) - 1
    for digit in digits:
        repr.append(digit * power ** i)
        i -= 1

    def convert(num):
        digits = []
        while num != 0:
            digits.append(str(num % power))
            num = num // power
        return "".join(digits)[::-1]

    return int(convert(sum(repr) // num2)), int((convert(sum(repr) % num2))) if convert(sum(repr) % num2) != '' else 0
```

```python
division_numeral(62543, 2, 8)
division_numeral(31261, 2, 8)
division_numeral(14530, 2, 8)
```

```python
from loguru import logger
```

```python
division_numeral(62543, 2, 8)
@logger.catch
def custom_convert_to(num, power1, power2):
    digits = []
    while num != 0:
        digits.append(str(division_numeral(num, power2, power1)[1]))
        num = division_numeral(num, power2, power1)[0]
    return ''.join(digits[::-1])
```

```python
custom_convert_to(62543, 8, 2)
```
