```python
def extract_name(name):
    letters = []
    for letter in name.lstrip().rstrip():
        if letter == '\n':
            break
        letters.append(letter)
    return ''.join(letters)
```

```python
def get_name(url):
    try:
        headers = {'Accept-Encoding': 'identity', 'User-agent': 'your bot 0.1'}
        html = rq.get(url, headers=headers)
    except HTTPError as e:
        return None
    try:
        bs = BeautifulSoup(html.text, 'html.parser')
        name = bs.find('h1').get_text()
    except AttributeError as e:
        return None
    return name
```

```python
def get_article(url):
    try:
        headers = {'Accept-Encoding': 'identity', 'User-agent': 'your bot 0.1'}
        html = rq.get(url, headers=headers)
    except HTTPError as e:
        return None
    try:
        bs = BeautifulSoup(html.text, 'html.parser')
        article = bs.find_all('div', {'class': "tn__content"})
    except AttributeError as e:
        return None
    return article[0] if len(article) != 0 else None
```

```python
def get_urls(url):
    links, n = [], []
    try:
        headers = {'Accept-Encoding': 'identity', 'User-agent': 'your bot 0.1'}
        html = rq.get(url, headers=headers)
    except HTTPError as e:
        return None
    try:
        bs = BeautifulSoup(html.text, 'html.parser')
        urls = bs.find_all('a', {'class': "alphabet-index__item-link"})
        for link in urls:
            links.append(link['href'])
            # n.append(get_name(link))
    except AttributeError as e:
        return None
    return links, n
```

```python
def get_cols_rows(a):
    row, cols_ = [], []
    for i in a.values():
        for col in i.keys():
            row.append(i[col])
            cols_.append(col)
    return row, cols_, list(a.keys())


def create_row(article):
    return pd.DataFrame(data=list(article.values())[0],
                        columns=get_cols_rows(article)[1],
                        index=get_cols_rows(article)[2])
```
